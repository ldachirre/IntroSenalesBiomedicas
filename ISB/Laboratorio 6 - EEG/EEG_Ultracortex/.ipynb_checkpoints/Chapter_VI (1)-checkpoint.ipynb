# Chapter VI: Fast Fourier Transform

## What is time complexity?
Time complexity is how much time does it take an algorithm to run as a function of the length of the input.

**Big-O notation**: it is a functional tool to determine the complexity of an algorithm. The letter O is used because the order of the function.

What about the particular case of the DFT?

$$X[k] = \sum_{n=0}^{N-1}x[n]W_N^{kn}$$
$$X[k]=\sum_{n=0}^{N-1}x[n]W_N^{kn}=x[0]W_N^0+x[1]W_N^{1k}+\dots+x[N-1]W_N^{(N-1)k}$$
The calculation of the direct DFT requires of N complex multiplication and ($N-1$) complex sums for each value. Considering there are N values of k a total number of complex operations is $N^2$ multiplications and $N\left(N-1\right)$ complex sums.

$$N(N)+N(N-1)=2N^2-N=O(N^2)$$
Complex multiplication requires 4 real multiplies and 2 real additions, whereas complex additions require just 2 real additions. Then, the $N^2$ complex multiplies are the primary concern.

### How can we obtain an efficient calculation of the DFT?
Decomposition of the calculation of the DFT with N points using multiple and smaller DFT.

Considering  a DFT with N points and discomposed in integer values: $N=L\cdot M$. In case  $x\left\lbrack n\right\rbrack$ does not have compund length (i.e. N prime number), it would be necessary to add zeros in the sequence (zero padding).

It is possible to put the N points of the sequence $x\left\lbrack n\right\rbrack$ in a bidimensional matrix of L rows and M columns. The same can be done with the calculated values of the DFT. The new array of the points of $x\left\lbrack n\right\rbrack$ as $X\left(k\right)$ can be row-major or col-major.


Assuming a col-major distribution for $x\left\lbrack n\right\rbrack$ in the matrix.

$$x[l, mL]:n=l+mL$$

Assuming a row-major distribution for $X\left\lbrack k\right\rbrack$ in the matrix.

$$X(p, q):k=Mp+q$$

Then, the direct DFT now is given by:

$$X(p, q)= \sum_{m=0}^{M-1}\sum_{l=0}^{L-1}x(l,m)W_N^{(Mp+q)(mL+l)}$$
Analyzing $W_N$ : 

$$W_N^{(Mp+q)(mL+l)}=W_N^{Mp(Lm)}W_N^{Mpl}W_N^{q(Lm)}W_N^{lq}$$
$$W_N^{Mp(Lm)}=W_N^{Npm}=1$$
$$W_N^{Mpl}=W_{N/M}^{pl}=W_L^{pl}$$
$$W_N^{qLm}=W_{N/L}^{mq}=W_M^{mq}$$
Simplifying:

$$X(p,q)=\sum_{l=0}^{L-1}$$
Then, it is possible to sub-divide the problem by using three steps:



* DFT of M points:

$$F(l,q)=\sum_{m=0}^{M-1}x(l,m)W_N^{mq};\quad q\in[0;M-1];\quad l\in[0;L-1]$$


* Argument of the exterior sum:

$$G(l, q)=W_N^{lq}F(l, q);\quad q\in[0,M-1];\quad l\in[0,L-1]$$


* DFT of L points:

$$X(p, q)=\sum_{l=0}^{L-1}G(l,q)W_L^{pl}$$
The computational cost would be lower than the direct DFT.



* **L** DFT transforms of **M** points ${\mathrm{LM}}^2$:  compound multiplications and $\mathrm{ML}\left(L-1\right)$  complex sums.

* **LM** complex multiplications

*  ${\mathrm{ML}}^2$ complex multiplications and $\mathrm{ML}\left(L-1\right)$ complex sums.

Total: $N\left(M+L+1\right)$ complex multiplications and $N\left(M+L-2\right)$ complex sums.

Example: $N=1000$  using L=2 and M=500 

* **Direct DFT**: ${\left(1000\right)}^2 ={10}^6$

* **Divide and conquer**: $1000\left(500+2+1\right)=50300$

**Example**

Calculate the 15 points DFT sequence: (L=5,M=3 )

![image.png](https://github.com/piero-latorre/DSP/blob/main/VI/image_1.png?raw=true)

```python
import numpy as np;
import matplotlib.pyplot as plt
M = 3;L = 5
x = np.arange(15).reshape((M, L)).T
x
```

```python

n = np.arange(M)
k = np.arange(M)

WN = np.exp(-1j * 2 * np.pi / M)
nk = np.outer(n, k)
WNnk = np.power(WN, nk)

X1 = np.dot(x[0, :], WNnk)
X2 = np.dot(x[1, :], WNnk)
X3 = np.dot(x[2, :], WNnk)
X4 = np.dot(x[3, :], WNnk)
X5 = np.dot(x[4, :], WNnk)

X3_dft = np.array([X1, X2, X3, X4, X5])

l = np.arange(L)
q = np.arange(M)

lq = np.outer(l, q)

W15 = np.exp(-1j * 2 * np.pi / (M * L))
W15lq = np.power(W15, lq)

X_dft = X3_dft * W15lq

n = np.arange(L)
k = np.arange(L)

WN = np.exp(-1j * 2 * np.pi / L)
nk = np.outer(n, k)
WNnk = np.power(WN, nk)

X1 = np.dot(X_dft[:, 0].T, WNnk)
X2 = np.dot(X_dft[:, 1].T, WNnk)
X3 = np.dot(X_dft[:, 2].T, WNnk)

X = np.array([X1, X2, X3]).T
X
```

```python
#   ya no  se tiene que usar la transpuestas :D
```

```python
X = X.reshape(-1)
print(np.abs(X))
plt.figure()
plt.stem(np.abs(X))
plt.xlabel("Samples (k)")
plt.ylabel("Magnitude")
plt.title("DFT using M=3 and L=5")
plt.show()
```

### Then, what is the Fast Fourier Transform?
FFT exploits symmetries of $e^{-j\frac{2\pi }{N}\mathrm{kn}} =W_N^{\mathrm{kn}}$

Periodicity in  $n,k$ $W_N^{\mathrm{kn}} =W_N^{k\left(N+n\right)} =W_N^{\left(k+N\right)n}$

Efficient calculation: Decimation in time building a big DFT from smaller ones.

Considering a signal of $x\left\lbrack n\right\rbrack$ with  $N=2^v$ elements. If $x\left\lbrack n\right\rbrack$ is not multiple of 2, it is possible to use **zero padding**.

Following the strategy "Divide and conquer": $M=\frac{\;N}{2}$, $L=2$. That means that  $x\left\lbrack n\right\rbrack$ is divided in two sequences:

$$f_1[n]=x[2n];\quad f_2=x[2n+1];\quad n \in \left[ 0;\frac{N}{2}-1 \right]$$
Given $f_1 \left\lbrack n\right\rbrack$ and $f_2 \left\lbrack n\right\rbrack$ are determinate by the decimation. The method is known as **temporal decimation algorithm (in time)**.

Separating $x\left\lbrack n\right\rbrack$ into even and odd-indexed sequences:

$$X[k]=\sum_{n=\text{even}}x[n]W_N^{kn}+\sum_{n=\text{odd}}x[n]W_N^{kn}$$
Substituting variables:

$$X[k]=\sum_{m=0}^{\frac{N}{2}-1}x[2m]W_N^{2mk}+\sum_{m=0}^{\frac{N}{2}-1}x[2m+1]W_N^{k(2m+1)$$
One important property is that

$$W_N^2=W_{\frac{N}{2}}$$
Then

$$X[k]=\sum_{m=0}^{\frac{N}{2}-1}x[2m]W_{\frac{N}{2}}^{mk}+W_N^{mk}\sum_{m=0}^{\frac{N}{2}-1}x[2m+1]W_{\frac{N}{2}}^{mk}$$
$$X[k]=X_{\text{even}}[k]+W_N^{mk}X_{\text{odd}}[k];\quad k\in[0;N-1];\quad \text{sum of}\space  2\times \frac{N}{2} \text{ point DFTs}$$
The basic block on each stage of the algorithm is to take two complex numbers $\left(a,b\right)$, multiply b by $W_N^r$, and then add and substract them from a. This gives as a result two complex numbers A and B. The basic block is called **twiddle**.

### What is the computational cost of FFT?
Keep splitting: each $\frac{N}{2}\to 2\frac{N}{4}$ DFTs

How many times? $\frac{N}{2},\frac{N}{4},\ldots,\frac{N}{2^{m-1} },\frac{N}{2^m }$

<!-- Applying the same idea to the two $\frac{N}{2}$-point DFTs, it can be done  times, so the total number of complex multiplies is approximately -->

$$\frac{N}{2}\log_2N$$
which is much smaller than .

**Example**

Use this method for calculating the DFT of a sinusoidal signal with frequency $\omega =\frac{\pi }{3}$ and compare the result with the fft function from scipy.

$$% Solution with MATLAB$$


```python
import numpy as np
import matplotlib.pyplot as plt

M = 30;L = 2

n = np.arange(60);x = np.sin(np.pi/3 * n)

X_new = np.fft.fft(x)

plt.figure(2)
plt.subplot(3, 1, 1);plt.stem(n, x, 'b');plt.xlabel('Samples (n)');plt.ylabel('Amplitude');plt.title('Input signal')

x = x.reshape((L, M))
n = np.arange(M)
k = np.arange(M)
WN = np.exp(-1j * 2 * np.pi / M)
nk = n.reshape(-1, 1) @ k.reshape(1, -1)
WNnk = WN ** nk
X1 = x[0] @ WNnk
X2 = x[1] @ WNnk
X3_dft = np.vstack((X1, X2))

l = np.arange(L)
q = np.arange(M)
lq = l.reshape(-1, 1) @ q.reshape(1, -1)
W = np.exp(-1j * 2 * np.pi / (M * L))
Wlq = W ** lq
X_dft = X3_dft * Wlq

n = np.arange(L)
k = np.arange(L)
WN = np.exp(-1j * 2 * np.pi / L)
nk = n.reshape(-1, 1) @ k.reshape(1, -1)
WNnk = WN ** nk
X = np.array([])

for i in range(M):
    X_aux = X_dft[:, i] @ WNnk
    X = np.concatenate((X, X_aux))

X = X.reshape((M,L)).ravel()

plt.subplot(3, 1, 2)
plt.stem(np.arange(len(X)), abs(X), 'r')
plt.xlabel('Samples (k)')
plt.ylabel('Magnitude')
plt.title('DFT using Radix-2')

plt.subplot(3, 1, 3)
plt.stem(np.arange(len(X_new)), abs(X_new), 'k')
plt.xlabel('Samples (k)')
plt.ylabel('Magnitude')
plt.title('DFT using FFT')

plt.tight_layout()
plt.show()
```

Given $N=2^v$, the decimation can be realized by $v=\log_2 N$ times. In this case, the complex multiplications are reduced to $\frac{N}{2}\log_2 N$ and the complex sum to ${\mathrm{Nlog}}_2 N$ .

Example

A microphone record an audio signal which contains the number 5 of a keyboard. This signal is denominated DTMF (Dual-Tone-Multi-Frequency). In this case, the frequencies are composed by:

![image.png](https://github.com/piero-latorre/DSP/blob/main/VI/image_8.png?raw=true)

Create the signals of the number 5 of the keyboard. Generate its frequency spectrum.

$$% Solution with MATLAB$$


```python
import numpy as np
import matplotlib.pyplot as plt

f1 = 770
f2 = 1336
fs = 8000
N = 8000
n = np.arange(N)
x = np.sin(2 * np.pi * f1 / fs * n) + np.sin(2 * np.pi * f2 / fs * n)

X = np.fft.fft(x)
k = n
f = k * fs / (N - 1) - fs / 2
f = np.fft.fftshift(f)

plt.figure(3)
plt.stem(f, abs(X), 'b')
plt.xlabel('Frequency (Hz)')
plt.ylabel('|X|')
plt.title('Frequency spectrum')
plt.show()
```
