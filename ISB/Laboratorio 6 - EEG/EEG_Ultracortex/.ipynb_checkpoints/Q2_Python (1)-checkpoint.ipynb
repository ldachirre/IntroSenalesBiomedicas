### **Question 1 (6.00 pts)**

In order to determine whether the audio file *chord-G.wav* is accurately labeled, follow these steps:

a) (1.00 pts) Start by reading the audio file chord-G.wav using the function `fs, y = wavfile.read("chord-G.wav")`. Next, create a `stem()` plot to visualize the signal. Be sure to properly label the axes and provide a suitable title for the plot.

```python
import scipy.io.wavfile as wavfile

import numpy as np
import pandas as pd 

import  matplotlib.pyplot as plt 

fs, y = wavfile.read("chord-G.wav")
y = y.astype(float)


plt.figure
plt.stem(y)
plt.title("Signal")
plt.xlabel("Frequency")
plt.ylabel("Range")
```

b) (1.00 pts) Now, you need to generate signals corresponding to the notes that make up a **G-chord**, which are **G, B, and D notes**. Create these signals and plot them for elements ranging from 0 to 1000. To display a specific range of elements on the plot, use the `set_xlim()` function. Keep in mind that the length of these vectors should be the same as the length of the original audio signal. These are the signals:

$$g = \sin(2\pi (196)t) + \sin(2\pi (392)t) \\
b = \sin(2\pi (247)t) + \sin(2\pi (494)t) \\
d = \sin(2\pi (147)t) + \sin(2\pi (294)t)$$

```python


t=np.arange(0,1000)
g=np.sin(2*(np.pi)*196*t)+np.sin(2*(np.pi)*392*t)
b=np.sin(2*(np.pi)*247*t)+np.sin(2*(np.pi)*494*t)
d=np.sin(2*(np.pi)*147*t)+np.sin(2*(np.pi)*294*t)

plt.figure
plt.grid('true')
plt.plot(t,g)

plt.plot(t,b)

plt.plot(t,d)

plt.xlim(500,1000)
```

c) (1.00 pts) Compute the **normalized cross-correlation** between the original audio signal and the three generated signals representing the **G-chord**. After calculating the cross-correlations, **sum the three maximum values** and **display** this result in your program.

```python


a1 = (y - np.mean(y)) / (np.std(y) * len(y))
a2 = (b - np.mean(b)) / (np.std(b))
a3 = (g - np.mean(g)) / (np.std(g))
a4 = (d - np.mean(d)) / (np.std(d))

cr1 = np.correlate(a1, a2, 'full')
cr2 = np.correlate(a1, a3, 'full')
cr3 = np.correlate(a1, a4, 'full')

s1=np.max(cr1)
s2=np.max(cr2)
s3=np.max(cr3)
 
print(s1+s2+s3)
```

d) (2.00 pts) Repeat the same process for a **D-chord**, which consists of the notes **D, F#, and A**. Generate signals for these notes and perform the same calculations as mentioned in step b) and c) These are the signals:

$$d = \sin(2\pi (147)t) + \sin(2\pi (294)t) \\
f_{sharp} = \sin(2\pi (185)t) + \sin(2\pi (370)t) \\
a = \sin(2\pi (220)t) + \sin(2\pi (440)t)$$

```python
t=np.arange(0,1000)

d=np.sin(2*(np.pi)*147*t)+np.sin(2*(np.pi)*294*t)
fsh=np.sin(2*(np.pi)*185*t)+np.sin(2*(np.pi)*370*t)
a=np.sin(2*(np.pi)*220*t)+np.sin(2*(np.pi)*440*t)

plt.figure
plt.grid('true')
plt.plot(t,d)
plt.plot(t,fsh)
plt.plot(t,a)

#plt.xlim(700,900)
```

$$low = \sin(2\pi (1000)t) \\

```python
t=np.arange(0,1000)

l=np.sin(2*(np.pi)*1000*t)
m=np.sin(2*(np.pi)*4000*t)
h=np.sin(2*(np.pi)*7000*t)

plt.figure
plt.title("Frecuencies of Audio Signal")
plt.grid('true')
plt.plot(t,l)

plt.figure
plt.grid('true')
plt.plot(t,m)

plt.figure
plt.grid('true')
plt.plot(t,h)


plt.xlabel("Time")
plt.ylabel("Frecuencies")
```

b) (1.00 pts) Compute the **normalized cross-correlation** between the original audio signal from **Question 1** and the three generated signals that represent the **low, mid, and high frequencies**. Display the **maximum values of the cross-correlations** in your program.

```python
a1 = (y - np.mean(y)) / (np.std(y) * len(y))
b2 = (l - np.mean(l)) / (np.std(l))
b3 = (m - np.mean(m)) / (np.std(m))
b4 = (h - np.mean(h)) / (np.std(h))

crr1 = np.correlate(a1, b2, 'full')
crr2 = np.correlate(a1, b3, 'full')
crr3 = np.correlate(a1, b4, 'full')

su1=np.max(crr1)
su2=np.max(crr2)
su3=np.max(crr3)
 
print(su1)
print(su2)
print(su3)
```

```python
n=np.arange(0,100)

a=[0.5095,-1.1090,0.5095]
b=[0,0.7619,-0.2761]

co=np.convolve(a,b)
co

plt.stem(co)
```
