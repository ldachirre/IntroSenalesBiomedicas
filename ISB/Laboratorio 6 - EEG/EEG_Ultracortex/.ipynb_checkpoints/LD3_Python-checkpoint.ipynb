```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from matplotlib import patches
from matplotlib.pyplot import axvline, axhline
from collections import defaultdict
from scipy.signal import lfilter, TransferFunction, residuez, tf2zpk

def zplane(z, p, filename=None):
    """Plot the complex z-plane given zeros and poles.
    """
    # get a figure/plot
    ax = plt.subplot(2, 2, 1)

    # Add unit circle and zero axes    
    unit_circle = patches.Circle((0,0), radius=1, fill=False,
                                 color='black', ls='solid', alpha=0.1)
    ax.add_patch(unit_circle)
    axvline(0, color='0.7')
    axhline(0, color='0.7')
    
    # Plot the poles and set marker properties
    poles = plt.plot(p.real, p.imag, 'x', markersize=9, alpha=0.5)
    
    # Plot the zeros and set marker properties
    zeros = plt.plot(z.real, z.imag,  'o', markersize=9, 
             color='none', alpha=0.5,
             markeredgecolor=poles[0].get_color(), # same color as poles
             )

    # Scale axes to fit
    r = 1.5 * np.amax(np.concatenate((abs(z), abs(p), [1])))
    plt.axis('scaled')
    plt.axis([-r, r, -r, r])

    """
    If there are multiple poles or zeros at the same point, put a 
    superscript next to them.
    """
    # Finding duplicates by same pixel coordinates (hacky for now):
    poles_xy = ax.transData.transform(np.vstack(poles[0].get_data()).T)
    zeros_xy = ax.transData.transform(np.vstack(zeros[0].get_data()).T)    

    # dict keys should be ints for matching, but coords should be floats for 
    # keeping location of text accurate while zooming

    # TODO make less hacky, reduce duplication of code
    d = defaultdict(int)
    coords = defaultdict(tuple)
    for xy in poles_xy:
        key = tuple(np.rint(xy).astype('int'))
        d[key] += 1
        coords[key] = xy
    for key, value in d.items():
        if value > 1:
            x, y = ax.transData.inverted().transform(coords[key])
            plt.text(x, y, 
                        r' ${}^{' + str(value) + '}$',
                        fontsize=13,
                        )

    d = defaultdict(int)
    coords = defaultdict(tuple)
    for xy in zeros_xy:
        key = tuple(np.rint(xy).astype('int'))
        d[key] += 1
        coords[key] = xy
    for key, value in d.items():
        if value > 1:
            x, y = ax.transData.inverted().transform(coords[key])
            plt.text(x, y, 
                        r' ${}^{' + str(value) + '}$',
                        fontsize=13,
                        )

    if filename is None:
        plt.show()
    else:
        plt.savefig(filename)
        print ('Pole-zero plot saved to ' + str(filename))
```

c) (1 pt) Find the transfer function H(z).

d) (2 pt) Find the zeros and poles of the function. What would be the ROC of the system (assume causality)? Is this system BIBO-stable? Justify if the system is IIR or FIR.

```python
import scipy.signal as signal
def loc_impz(g,f,n):
    j = np.zeros(n+1)
    j[0] = 1
    h = signal.lfilter(f,g,j)
    return h
```

```python
import scipy.signal as sig
(a,b,c,d,e,f) = 0.29289322, -0.58578644, 0.29289322, 1, 0, -0.171572875
n = np.arange(0, 11)


g = np.array([1, -d*e, -d*f]) #a
f = np.array([d*a, d*b, d*c]) #b

z, p, k = sig.tf2zpk(f, g)
plt.figure(figsize=(20, 12))
plt.subplot(2, 2, 1)
plt.grid(True, color='0.9', linestyle='-', which='both', axis='both')
plt.title('Poles and zeros')
zplane(z, p)

#If we see the poles we can see, the ROC is 1 and it's FIR. and if we see the impulse response we can conclude it´s BIBOestable
```

e) (4 pt) Find the impulse response of the filter. From the impulse response, obtain the magnitude and phase of the frequency response of the filter. 

Use 128 points and set the x axis in Hz. What type of filter is it (high-pass, low-pass, band-pass)?

```python
# Compute the impulse response
n = np.arange(128)
y=loc_impz(g,f,len(n)-1)
# Plot the impulse response
plt.figure()
plt.stem(n,y)
plt.xlabel('Samples (n)')
plt.ylabel('Amplitude')
plt.title('Impulse Response')
plt.show()

#It´s a low-pass filter
```

```python
import scipy.fft as ft
def fft_fast(x,fs):
  X = ft.fftshift(ft.fft(x))
  N = len(x)
  k = np.arange(N)
  w = 2*np.pi*k/N
  w = ft.fftshift(w)
  w = np.unwrap(w - 2*np.pi)
  f = w*fs/(2*np.pi)
  return X,f,w

[H, fh, wh]= fft_fast(x, 128) 

fig, ax = plt.subplots(figsize=(10, 2.5))
plt.plot(wh, abs(H),"lightcoral"); plt.title('H filter by FFT')
plt.xlabel('Frequency (rad/s)'); plt.ylabel('Magnitude'); 
ax.set_xticks([-np.pi, -2*np.pi/3, -np.pi/3, 0, np.pi/3, 2*np.pi/3, np.pi]); 
labels = [r'$-\pi$', r'$-2\pi/3$', r'$\pi/3$', r'$0$', r'$\pi/3$', r'$2\pi/3$', r'$\pi$']
ax.set_xticklabels(labels)
plt.grid(True)

fig, ax = plt.subplots(figsize=(10, 2.5))
plt.plot(wh, np.unwrap(np.angle(H)),"lightcoral"); plt.title('H filter by FFT')
plt.xlabel('Frequency (rad/s)'); plt.ylabel('Phase'); 
ax.set_xticks([-np.pi, -2*np.pi/3, -np.pi/3, 0, np.pi/3, 2*np.pi/3, np.pi]); 
labels = [r'$-\pi$', r'$-2\pi/3$', r'$\pi/3$', r'$0$', r'$\pi/3$', r'$2\pi/3$', r'$\pi$']
ax.set_xticklabels(labels)
plt.grid(True)
```

```python
#It´s a frecuency shift,can be used to filter out the noise because 
#the noise is often spread out over a wide range of frequencies, whereas the signal of 
#interest is usually concentrated around a specific frequency or frequency range. By shifting 
#the frequency spectrum of the signal so that the signal's frequency of interest is centered at zero frequency, 
#it is easier to isolate the signal from the noise.
```
