```python
import numpy as np
import matplotlib.pyplot as plt


N = 8*np.lcm(6, 4)  # Ideal number of elements

# 2)
n = np.arange(N)
# k = np.arange(N)
x_s = 3/8 * np.cos(10/30* np.pi * n)+ 4/8 *np.cos(6/12 * np.pi* n)+ 5/8 *np.cos(14/28 * np.pi *n) 

X_s, ff = np.fft.fft(x_s, norm="ortho"), np.fft.fftfreq(N)

# Graphics
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 8))

ax1.plot(ff, np.abs(X_s))
ax1.set_title("DFT of X[n], Magnitude and Phase Spectrum")
ax1.set_ylabel("Magnitude")
ax1.set_xlabel("Frequency (from -Ï to Ï)")     #remember that we plot signal in range of -pi , pi

ax2.plot(ff, np.angle(X_s))
ax2.set_ylabel("Phase (rad)")
ax2.set_xlabel("Frequency (from -Ï to Ï)")

plt.tight_layout()
plt.show()
```

2. Find the inverse DFT for the given signal. Plot the signal with the appropiate labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function x[n]
n = np.arange(0, 100)
x = 3/8 * np.cos(10/30 * np.pi * n) + 4/8 * np.cos(6/12 * np.pi * n) + 5/8 * np.cos(14/28 * np.pi * n)

# Calculate the DFT of x[n]
X = np.fft.fft(x)

# Calculate the inverse DFT of X[k]
x_inv = np.fft.ifft(X)

# Plot the original signal and its inverse DFT
plt.figure(figsize=(8,6))
plt.stem(n, X, linefmt='C0-', markerfmt='C0o', label='DFT')
plt.stem(n, x_inv.real, linefmt='C1-', markerfmt='C1x', label='Inverse DFT')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('DFT and its inverse DFT')
plt.legend()
plt.show()
```
