```python
import numpy as np
import scipy.signal as sig
import scipy.fft as ft
def fundamental_range_freq(signal):
  N_size = len(signal);
  w = 2*np.pi* np.arange(N_size-1)/ N_size;
  w = ft.fftshift(w);
  w = np.unwrap(w - 2*np.pi);
  return w
def radian_range_freq(signal,N):
    N_size = len(signal);
    w = N*np.pi*np.arange(N_size-1)/ N_size;
    w = ft.fftshift(w);
    w = np.unwrap(w - N*np.pi);
    return w
```

## What is the Discrete-Time Fourier Transform?
Frequency analysis of discrete-time provides an important analysis and design tool into the solution of problems.

The Discrete-Time Fourier Transform (DTFT) of a signal  is the complex-valued of the **continuous** (frequency) variable $\omega$ defined by:
$$X(e^{j\omega}) = \sum_{n = -\infty}^\infty x[n] e^{-jn\omega}$$
The DTFT is the representation of a discrete-time periodic and aperiodic signal as a weighted sum of complex exponentials. This operation also means that the sum in the previous equation must converge.

A sufficient condition for the sum to converge uniformly to a continuous function of $\omega$ is that $x[n]$ is absolutely summable.


```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib widget

# Compute the DTFT of a sinusoidal using for loops
n = np.arange(-100,101)
f_rad=np.pi/6
x = np.sin(f_rad * n)
w = fundamental_range_freq(x)
X = np.zeros_like(w, dtype=complex)

for i in range(len(w)):
    for j in range(len(n)):
        X[i] += x[j] * np.exp(-1j * w[i] * n[j])

# Plot the original signal and the DTFT
fig, axs = plt.subplots(2, 1)
fig.suptitle('DTFT of a Sinusoidal Signal')
axs[0].stem(n, x, 'b');axs[0].set_xlabel('Samples (n)');axs[0].set_ylabel('Amplitude')
axs[0].set_title('Original signal');axs[0].grid(True)

axs[1].plot(w/np.pi, np.abs(X), 'r');axs[1].set_xlabel('Frequency in pi units');axs[1].set_ylabel('Magnitude')
axs[1].set_title('DTFT of the signal');axs[1].grid(True)

plt.tight_layout()
```

```python
find_max= np.argmax(abs(X))
max_freq=abs(w[find_max])

print("The value of the sin was %f and the maximum frequency according to the DTFT is %f \n"%(f_rad,max_freq));
```


The DTFT is a complex-number operation with some interesting properties. Since $e^{-\mathrm{jn}\omega }$ is periodic in $\omega$ with period $2\pi$ means that $X\left(e^{j\omega } \right)=X\left(e^{j\left(\omega +2\pi \right)} \right)$.


```python

w = radian_range_freq(x,6)
X_ejw = np.zeros_like(w,dtype=complex)


for i in range(len(w)):
    for j in range(len(n)):
        X_ejw[i] += x[j] * np.exp(-1j * w[i] * n[j])

fig, (ax1, ax2) = plt.subplots(2, 1)

ax1.stem(n, x, 'b')
ax1.set(title='x[n]', xlabel='Samples [n]', ylabel='Amplitude')


ax2.plot(w, abs(X_ejw), 'r')
ax2.set(title='X(e^{jw})', xlabel='Frequency [rad]', ylabel='Magnitude')

plt.tight_layout()
```

## How can we calculate the inverse of the DTFT?
The inverse of the DTFT also can be calculated by using the definition:
$$x[n] = \frac{1}{2\pi} \int_{-\pi}^\pi X(e^{j\omega})e^{jn\omega} d\omega$$
**Example**

Assume a finite energy signal :
$$X(\omega) = \begin{cases} 1,\quad |\omega| \leq \omega_c \\ 0,\quad \omega_c \leq |\omega| \leq \pi \end{cases}$$
Find $x\left\lbrack n\right\rbrack$:


```python

n = np.arange(-80, 81)
wc = np.pi/6
x = wc/np.pi*np.sinc(wc/np.pi*n)

w = fundamental_range_freq(n)
X = x @ np.exp(-1j*np.outer(n,w))

fig, axs = plt.subplots(2, 1, figsize=(8, 8))

axs[0].stem(n, x, 'b')
axs[0].set_xlabel('Samples (n)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Original signal')
axs[0].grid(True)

X_mag = np.abs(X)
axs[1].plot(w/np.pi, X_mag, 'r')
axs[1].set_xlabel('Frequency in pi units')
axs[1].set_ylabel('Magnitude')
axs[1].set_title('DTFT of the signal')
axs[1].grid(True)
plt.tight_layout()

```

## What is the relation between DTFT and Z-Transform?
Given a sequence $x\left\lbrack n\right\rbrack$ and the complex variable $z={\mathrm{re}}^{j\omega }$

$$X(z)|_{z=re^{j\omega}} = \sum_{n=-\infty}^\infty (x[n]r^{-n})e^{-j\omega n}$$
Then, $X\left(z\right)$ can be interpreted as the Fourier Transform in Dicrete-Time of $x\left\lbrack n\right\rbrack r^{-n}$.

If $X\left(z\right)$ converges using $|z|=1$, the DTFT corresponds to the Z-transform evaluated in the unit circle $r=1$ .
$$X(z)|_{z=e^{j\omega}} = \sum_{n=-\infty}^\infty x[n] e^{-j\omega n}$$
If  does not include $|z|=1$ in its region of convergence, then  does not exist.

### Convolution

![image.png](https://github.com/piero-latorre/DSP/blob/main/IV/image_4.png?raw=true)


```python

n = np.arange(-100, 101)
x = np.sin(np.pi/6*n)
h = (np.pi/12)/np.pi*np.sinc((np.pi/12)/np.pi*n)
y = np.convolve(x, h)
ny = np.arange(2*n[0], 2*n[-1]+1)
M = 500;k = np.arange(-M, M+1);w = np.pi/500 * k

X = x @  np.exp(-1j*np.pi/M)**(np.outer(n, k))
H = h @  np.exp(-1j*np.pi/M)**(np.outer(n, k))
Y = y @  np.exp(-1j*np.pi/M)**(np.outer(ny, k))

plt.figure(8, figsize=(18, 8))
plt.subplots_adjust(hspace=0.5, wspace=0.3)

plt.subplot(3, 3, 1)
plt.stem(n, x, linefmt='b-', basefmt='b:');plt.xlabel('Samples (n)');plt.ylabel('Amplitude')
plt.title('x[n] signal');plt.grid(True)

plt.subplot(3, 3, 2);plt.stem(n, h, linefmt='b-', basefmt='b:')
plt.xlabel('Samples (n)');plt.ylabel('Amplitude')
plt.title('h[n] signal');plt.grid(True)

plt.subplot(3, 3, 3);plt.stem(ny, y, linefmt='b-', basefmt='b:')
plt.xlabel('Samples (n)')
plt.ylabel('Amplitude')
plt.title('y[n] signal')
plt.grid(True)

plt.subplot(3, 3, 4)
plt.plot(w/np.pi, np.abs(X), 'r')
plt.ylabel('Magnitude')
plt.title('DTFT magnitude of x[n]')
plt.grid(True)

plt.subplot(3, 3, 5)
plt.plot(w/np.pi, np.abs(H), 'r')
plt.ylabel('Magnitude')
plt.title('DTFT magnitude of h[n]')
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(w/np.pi, np.abs(Y), 'r')
plt.ylabel('Magnitude')
plt.title('DTFT magnitude of y[n]')
plt.ylim([0, 100])
plt.grid(True)

plt.subplot(3, 3, 7)
plt.plot(w/np.pi, np.unwrap(np.angle(X)), 'k')
plt.xlabel('Frequency in π units')
plt.ylabel('Phase')
plt.title('DTFT phase of x[n]')
plt.grid(True)

plt.subplot(3, 3, 8)
plt.plot(w/np.pi, np.unwrap(np.angle(H)), 'k')
plt.xlabel('Frequency in π units')
plt.ylabel('Phase')
plt.title('DTFT phase of h[n]')
plt.grid(True)

plt.subplot(3, 3, 9)
plt.plot(w/np.pi, np.unwrap(np.angle(Y)), 'k')
plt.xlabel('Frequency in π units')
plt.ylabel('Phase')
plt.title('DTFT phase of y[n]')
plt.grid(True)

plt.tight_layout()
```

## How can we apply the DTFT fot LTI Systems?
One of the most important properties of the DTFT is the direct application with the convolution:
$$y[n]=x[n]*h[n]$$
And the equality with the product of the transforms:
$$Y(e^{j\omega}) = X(e^{j\omega}) H(e^{j\omega})$$
**Example**

Given the system defined by the following difference equations:
$$w[n]=x[n]-0.81w[n-2]$$
$$y[n]=w[n]+w[n-1]$$
* Find the transfer function

* Find the output of the system when  and  if the input is

$$x[n] = \begin{cases} 1,\quad \text{if n even} \\ 0, \quad \text{if n odd} \end{cases}$$


```python

b = np.array([1, 1]);a = np.array([1, 0, 0.81])
n = np.arange(-100, 101);x = np.zeros(len(n))
x[np.where(n % 2 == 0)] = 1

y =sig.lfilter(b,a,x)

M = 500;k = np.arange(-M, M+1);w = (np.pi/500)*k

X=x @  np.exp(-1j*np.pi/M)**(np.outer(n, k))

Y =y @  np.exp(-1j*np.pi/M)**(np.outer(n, k))

plt.figure(figsize=(20,12))

plt.subplot(3, 1, 1)
plt.stem(n, x, 'b')
plt.xlabel('Samples (n)')
plt.title('x signal')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(w/np.pi, np.abs(X), 'r')
plt.title('DTFT of x signal')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(w/np.pi, np.abs(Y), 'k')
plt.xlabel('Frequency in \pi units')
plt.title('DTFT of y signal')
plt.grid(True)


plt.tight_layout()
```
