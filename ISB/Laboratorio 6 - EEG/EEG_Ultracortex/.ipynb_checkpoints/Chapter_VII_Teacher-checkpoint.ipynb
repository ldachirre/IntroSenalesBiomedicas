### **What is the procedure to obtain a discrete signal from a continuous signal?**

1. **Modulation by an impulse train**: 
given an impulse train $s(t)$:

$$s(t) = \sum_{n=-\infty}^\infty \delta(t-nT)$$

A modulated signal is obtained $x_s(t)$:

$$x_s(t)=x_c(t)s(t)=x_c(t)\sum_{n=-\infty}^\infty\delta(t-nT)=\sum_{n=-\infty}^\infty x_c(nT)\delta(t-nT)$$

The Fourier Transform of the impulse train is defined as:

$$s(t)\leftrightarrow S(j\Omega)=\frac{2\pi}{T}\sum_{k=-\infty}^\infty\delta(\Omega-k\Omega_s),\quad \Omega_s=\frac{2\pi}{T}$$
Applying the multiplication property:

$$x_s(t)=x_c(t)s(t)\leftrightarrow X_s(j\Omega)=\frac{1}{2\pi}X_c(j\Omega)*S(j\Omega)$$
$$X_s(j\Omega)=\frac{1}{T}\sum_{k=-\infty}^\infty X_c(j(\Omega-k\Omega_s))$$
Hence, $X_s(j\Omega)$ is a periodic repetition of $X_c(j\Omega)$ shifted by $\Omega_s$  multiples.

Assuming  real, there is a minimum value of  to avoid aliasing.

2. **Impulse train to a discrete sequence:** $x[n]$  is obtained with a simple discrete domain with no information about sample period.

3. **Frequency normalization:** while the spectrum of $x_s(t)$ is in function of $\Omega$, the spectrum of $x[n]$ use a normalized frequency $\omega = \Omega T = \frac{\Omega}{F_s}$

$$x[n]\leftrightarrow X(e^{j\omega})=X_s(j\Omega)|_{\Omega=\frac{\omega}{T}}$$
$$X(e^{j\omega})=\frac{1}{T}\sum_{k=-\infty}^\infty X_c\left(j\left(\frac{\omega}{T}-k\frac{2\pi}{T}\right) \right)$$
4. **Effect of frequency normalization:** $\omega = 2\pi$ represents $\Omega = \Omega_s$



```python
import numpy as np
import matplotlib.pyplot as plt

# Define time vector and signal
t = np.arange(0, 5, 0.0001);xc = np.sin(2*np.pi*10*t) + 2*np.sin(2*np.pi*70*t)

# Compute DFT
Wmax = 2*np.pi*100;K = 500;k = np.arange(K+1);W = k*Wmax/K
Xa = np.dot(xc, np.exp(-1j*np.outer(t, W))) * (1/len(t))
Xa = np.abs(Xa)

# Plot signal and spectrum
W = np.concatenate((-np.flipud(W), W[1:]))
Xa = np.concatenate((np.flipud(Xa), Xa[1:]))

fig, (ax1, ax2) = plt.subplots(2, 1,figsize=(7,5))
ax1.plot(t, xc, 'b')
ax1.set_xlabel('Time (s)')
ax1.set_title('xc(t) signal')
ax1.set_xlim([0, 1])
ax1.grid(True)
ax2.plot(W/(2*np.pi), Xa, 'r')
ax2.set_xlabel('Frequency (Hz)')
ax2.set_title('Xc(jΩ)')
ax2.grid(True)
plt.show()
```

```python
s = np.zeros((len(t),))
s[::100] = np.double(1.0)

Wmax = 2*np.pi*100;K = 500;k = np.arange(0, K+1);W = k*Wmax/K

# calculate Fourier transform
Xs = np.dot(s, np.exp(-1j*np.outer(t, W))) * (1/len(t))
Xs = np.abs(Xs)

W = np.concatenate((-np.flip(W), W[1:]))
Xs = np.concatenate((np.flip(Xs), Xs[1:]))

# plot results
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(t, s, 'b')
ax1.set_xlabel('Time (s)')
ax1.set_title('s(t) signal')
ax1.set_xlim([0.2, 0.4])
ax1.grid(True)

ax2.plot(W/(2*np.pi), Xs, 'r')
ax2.set_xlabel('Frequency (Hz)')
ax2.set_title('S(jw)')
ax2.grid(True)
```

```python
xs1 = xc * s
Wmax = 2 * np.pi * 100;K = 500;k = np.arange(K+1)
W = k * Wmax / K

exp_term = np.exp(-1j * np.outer(t, W))
X = xs1 @ exp_term * (1/len(t))
X = np.abs(X)

W = np.concatenate((-np.flip(W), W[1:]))
X = np.concatenate((np.flip(X), X[1:]))

plt.figure(3)
plt.subplot(2, 1, 1)
plt.plot(t, xs1, 'b')
plt.xlabel('Time (s)')
plt.title('x_s(t) signal')
plt.xlim([0, 1])
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(W/(2*np.pi), X, 'r')
plt.xlabel('Frequency (Hz)')
plt.title('X_s(jΩ)')
plt.grid(True)
```

### **What is the procedure for reconstruction of discrete signals?**
Given a signal $x_c(t)$ with spectrum $x_c(j\Omega) = 0, |\Omega| \leq \Omega_{max}$ if there is a sampled signal $x[n]$ with $F_s\geq 2\Omega_{max}$ , then the repetition in the sampled signal will not generate aliasing.

Additionally, the spectrum of the sampled signal $X(e^{j\omega})$ is identical to the spectrum analog signal $X(j\Omega)$ in the fundamental range frequency $-\pi\leq\omega\leq\pi$.

On the other hand, if $F_s \leq 2\Omega_{max}$ an overlap in the fundamental rangre frequency will be generated.

Without aliasing in the discrete version $x[n]$, it is possible to reconstruct $x_c(t)$

Summary:

* Spectrum of $x_c(t)\rightarrow X(j\Omega)$

* Spectrum of $x[n] \rightarrow X(e^{j\omega})$

Conversion of discrete sequence to impulse train:

$$x_s(t)=\sum_{n=-\infty}^\infty x[n]\delta (t-nT)$$
Convolution with ideal interpolation filter

$$h_r(t)=\frac{\sin\left(\frac{\pi}{T}t\right)}{\frac{\pi}{T}t};\quad x_r(t)=\sum_{-\infty}^\infty x[n]h_r(t-nT)=\sum_{-\infty}^\infty x_c(nT)\frac{\sin\left(\frac{\pi(t-nT)}{T}\right)}{\frac{\pi(t-nT)}{T}}$$
The spectrum of the system $H(j\Omega)$ corresponds to an ideal low-filter with gain $T$ and with a cut frequency of $\frac{\pi}{T}$:

$$H_r(j\Omega)=\begin{cases} T,\quad |\Omega| \leq \frac{\pi}{T} \\ 0,\quad \text{otherwise} \end{cases}$$
Then, $h_r(t)$ is a continuous system able to avoid replicas out of the fundamental range.



```python
# Define time vector
t = np.arange(0, 5, 0.01)

# Generate aliased signal
nT = np.arange(0, 5, 0.01)
n = nT / 0.01
xs1 = np.sin(2 * np.pi * 10 / 100 * n) + 2 * np.sin(2 * np.pi * 70 / 100 * n)

# Reconstruct aliased signal
y1 = np.zeros_like(t)
for i in range(len(t)):
    for j in range(len(n)):
        y1[i] += xs1[j] * np.sinc((t[i] - nT[j]) / 0.01)

# Plot reconstructed aliased signal
plt.figure()
plt.plot(t, y1, 'b')
plt.xlabel('Time (s)')
plt.title('Reconstruction of aliased signal')
plt.xlim([0, 1])
plt.grid(True)

# Generate non-aliased signal
xs2 = np.sin(2 * np.pi * 10 / 100 * n)

# Reconstruct non-aliased signal
y2 = np.zeros_like(t)
for i in range(len(t)):
    for j in range(len(n)):
        y2[i] += xs2[j] * np.sinc((t[i] - nT[j]) / 0.01)

# Plot reconstructed non-aliased signal
plt.figure()
plt.plot(t, y2, 'b')
plt.xlabel('Time (s)')
plt.title('Reconstruction of non-aliased signal')
plt.xlim([0, 1])
plt.grid(True)

plt.show()
```

### **How can we convert the sampling rate?**
Given a discrete signal $x[n]$ sampled with $F_x=\frac{1}{T_x}$ : is it possible to change its sampling frequency using discrete-time operation?

* **Alternative 1**: $x[n]$ must be converted to $y_a(t)$ . Then, using a D/A take the samples of the result with $F_y=\frac{1}{T_y}$ . Given the reconstruction filter $g(t)$ and the modulated impulse train $\hat{x}_s(t)$, the reconstructed signal  $\hat{x}(t)$ corresponds to:

$$\hat{x}=\left(\sum_{\hat{n}=-\infty}^\infty x(\hat{n}T_x)\delta(t-\hat{n}T_x)\right)*g(t)$$
$$=\sum_{\hat{n}=-\infty}^\infty x(\hat{n}T_x)g(t-\hat{n}T_x)$$
After that, the sequence obtained under the new sampling frequency corresponds to:

$$y[n]=\hat{x}(nT_y)\sum_{\hat{n}=-\infty}^\infty x(\hat{n}T_x)g(nT_y-\hat{n}T_x)$$
This allows to obtain $y[n]$ sampled at $F_y=\frac{1}{T_y}$ from $x[n]$ sampled at $F_x=\frac{1}{T_x}$.

* **Alternative 2**: entirely discrete processing

**Downsampling:** reduce the sampling frequency $F_x$ by an integer factor of $D$.

![](https://github.com/piero-latorre/DSP/blob/main/VII/image_7.png?raw=true)

**Upsampling:** increase the sampling frequency $F_x$ by an integer factor of $I$.

###**What is the process for downsampling?**

Reduce $F_x$ by an integer factor of $D$. From (*) and the relation between $T_y = DT_x$ :

$$y(nT_y)=\sum_{\hat{n}=-\infty}^\infty x(\hat{n}T_x)g(nDT_x-\hat{n}T_x)$$
Considering the ideal interpolated filter $g(t)$ scrolls in increments values of $DT_x$ , $y(nT_y)$  takes one of each $D$ samples of $x(nT_x)$. Then, the relation between $y[n]$ and $x[n]$ is $y[n]=x[nD]$.



```python
# Given a signal with sampling frequency of 100 Hz, downsample it to 50 Hz
fs = 100;ts = 1/fs;
nT = np.arange(0, 1, ts)
nx = np.arange(0, len(nT))
x = np.sin(2*np.pi*10*nT)
D = 2
x_down = x[::D]
nx_down = np.arange(0, len(x_down))

# Plot the original and downsampled signals
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.stem(nx, x, linefmt="b-", markerfmt="bo", basefmt=" ")
ax1.set_title("Fs = 100 Hz")
ax2.stem(nx_down, x_down, linefmt="r-", markerfmt="ro", basefmt=" ")
ax2.set_xlabel("Samples (n)")
ax2.set_title("Fs = 50 Hz")
plt.tight_layout()
```

The subsample $y[n]$ is characterized by the frequency spectrum:

$$y[n]=x[nD]\leftrightarrow Y(e^{j\omega})=\frac{1}{D}\sum_{i=0}^{D-1}X\left(e^{j\left(\frac{\omega}{D}-\frac{2\pi i}{D}\right)}\right)$$
It is possible that reducing the sampling frequency to $\frac{F_x}{D}$ generates aliasing. Considering that angular frequency domain $\Omega$ of $x_c(t)$ and the normalized frequency domain $\omega_x=\Omega T_x$ of the sequence $x[n]$ , the normalized frequency domain $\omega_y$ of the sequence $y[n]$ corresponds to:

$$\omega_y = \Omega T_y$$
$$\omega_y = \Omega(DT_x)$$
$$\omega_y=D\omega_x$$
Hence, the information of $y[n]$ is included in the fundamental range $(|\omega_y|\leq\pi)$ if the information of $x[n]$ is included in $|\omega_x|\leq\frac{\pi}{D}$.



```python
import numpy as np
import matplotlib.pyplot as plt

# compute and plot FFT of signals
X = np.fft.fftshift(np.fft.fft(x))
X_down = np.fft.fftshift(np.fft.fft(x_down))


w=2*np.pi*nx/len(nx);
w=np.fft.fftshift(w);
w=np.unwrap(w-2*np.pi)
w_down=2*np.pi*nx_down/len(nx_down);w_down=np.fft.fftshift(w_down);w_down=np.unwrap(w_down-2*np.pi)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(w*100/2/np.pi, np.abs(X)/len(X))#, use_line_collection=True)
axs[0].set(title='FFT of original signal', ylabel='Magnitude')
axs[1].stem(w_down*50/2/np.pi, np.abs(X_down)/len(X_down))#, use_line_collection=True)
axs[1].set(title=f'FFT of downsampled signal (Fs = {50} Hz)', xlabel='Frequency (Hz)', ylabel='Magnitude')
fig.tight_layout()
```

```python

# Define time vector
t = np.arange(0, 5, 0.0001)

# Define signals
xc = np.sin(2*np.pi*10*t)
s = np.zeros((len(t),))
s[::100] = np.double(1.0)
xs1 = xc * s

# Compute Fourier transform
Wmax = 2*np.pi*100
K = 500
k = np.arange(K+1)
W = k*Wmax/K
X = xs1 @ np.exp(-1j*np.outer(t,W)) * (1/len(t))
X = np.abs(X)

# Plot signals and Fourier transform
W = np.concatenate((-np.flipud(W), W[1:]))
X = np.concatenate((np.flipud(X), X[1:]))
fig, axs = plt.subplots(2,1)
fig.suptitle('Signal and Fourier Transform')
axs[0].plot(t, xs1, 'b')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('xs(t) signal')
axs[0].set_xlim([0,1])
axs[0].grid(True)
axs[1].plot(W/(2*np.pi), X, 'r')
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Xs(jΩ)')
axs[1].grid(True)
plt.show()
```

###**The obtained signal does not represent the original signal, what solution do we have?**

The process of limiting the bandwith of $x[n]$ to $|\omega_x|=\frac{\pi}{D}$ and then apply a downsampling by a factor of $D$ it is known as **decimation**.

Decimation process:

$$v[n]=x[n]*h_D[n]\leftrightarrow X(e^{j\omega})H_D(e^{j\omega}),\quad H_D(e^{j\omega})=\begin{cases}1,\quad |\omega|\leq \frac{\pi}{D}\\0,\quad \text{otherwise}\end{cases}$$
$$y[n]=v[nD]$$
From the relation $\omega_y=D\omega_x$, it is known that the interval $|\omega_x|\leq\frac{\pi}{D}$ is projected as $|\omega_y|\leq\pi$.

![](https://github.com/piero-latorre/DSP/blob/main/VII/image_9.png?raw=true)


```python
from scipy.fft import fftshift, fft
from scipy.signal import decimate
# Generate the signal
fs = 100;ts = 1/fs;nT = np.arange(0, 1, ts)
nx = np.arange(len(nT))
x = np.sin(2*np.pi*40*nT)

# Apply decimation
D = 2
x_down =decimate(x,D)
nx_down = np.arange(len(x_down))

# Plot the original and decimated signals
fig, axs = plt.subplots(2, 1)
axs[0].stem(nx, x, use_line_collection=True)
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(nx_down, x_down, use_line_collection=True)
axs[1].set_xlabel('Samples (n)')
axs[1].set_title('Fs = 50 Hz')

# Compute and plot the FFTs
X = fftshift(fft(x))
X_down = fftshift(fft(x_down))

w = 2*np.pi*nx/len(nx)
w = np.fft.fftshift(w)
w = np.unwrap(w - 2*np.pi)

w_down = 2*np.pi*nx_down/len(nx_down)
w_down = np.fft.fftshift(w_down)
w_down = np.unwrap(w_down - 2*np.pi)
fig, axs = plt.subplots(2, 1)
axs[0].stem(w*fs/(2*np.pi), np.abs(X)/len(X), use_line_collection=True)
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(w_down*fs/(2*np.pi), np.abs(X_down)/len(X_down), use_line_collection=True)
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_title('Fs = 50 Hz')
plt.tight_layout()
```

###**What is the process for upsampling?**

Increase $F_s$ by an integer factor of $I$. From (*) and the relation between $T_y=\frac{T_x}{I}$:

$$y(nT_y)=\sum_{\hat{n}=-\infty}^\infty x(\hat{n}T_x)g\left(n\frac{T_x}{I}-\hat{n}T_x\right)$$
Since $g(t)$ moves in increments of $\frac{T_x}{I}$, $y(nT_y)$ includes $I-1$ new samples between each element of $x(nT_x)$. Then, the following equation is formulated:

$$y[n]=\begin{cases}x\left[\frac{n}{I}\right],\quad n=rI,\quad r\in\mathbb{Z} \\ 0,\quad \text{otherwise}\end{cases}$$


```python
fs = 100
ts = 1/fs
nT = np.arange(0, 1, ts)
nx = np.arange(0, len(nT))
x = np.sin(2*np.pi*10*nT)

I = 2
x_up = np.zeros(I*len(x))
x_up[::I] = x

nx_up = np.arange(0, len(x_up))
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(nx, x, use_line_collection=True)
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(nx_up, x_up, use_line_collection=True)
axs[1].set_xlabel('Samples (n)')
axs[1].set_title('Fs = 200 Hz')
plt.show()
```

The upsample sequence $y[n]$ is characterized by the frequency spectrum:

$$Y(e^{j\omega})=X(e^{j\omega I})$$
Given a  $X(e^{j\omega})$ presents repetitions every $2\pi$, $Y(e^{j\omega})$ includes repetition within the fundamental range $I\geq2$. Considering that angular frequency domain $\Omega$ of the signal $x_c(t)$ and the normalized $\omega_x=\Omega T_x$ of the sequence $x[n]$, then the normalized frequency domain $\omega_y$ of the sequence $y[n]$ corresponds to

$$\omega_y=\Omega T_y$$
$$\omega_y=\Omega \frac{T_x}{I}$$
$$\omega_y=\frac{\omega_x}{I}$$
Hence, the information in the frequency of $y[n]$ in the fundamental range $(|\omega_y|\leq\pi)$ includes the content of the frequency of $x[n]$ in $(|\omega_x\leq\pi|)$.



```python
# Given a signal with sampling frequency of 100 Hz, upsample it to 200 Hz
fs = 100
ts = 1/fs
nT = np.arange(0, 1, ts)
nx = np.arange(0, len(nT))
x = np.sin(2*np.pi*10*nT)
I = 2
x_up = np.zeros(len(x)*I)
x_up[::I] = x
nx_up = np.arange(0, len(x_up))

# Compute the FFT of the previous signals
X = np.fft.fftshift(np.fft.fft(x))
X_up = np.fft.fftshift(np.fft.fft(x_up))
w = np.fft.fftshift(np.linspace(-np.pi, np.pi, len(nx)))
w_up = np.fft.fftshift(np.linspace(-np.pi, np.pi, len(nx_up)))

# Plot the original and upsampled signals
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(nx, x)
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(nx_up, x_up, 'r')
axs[1].set_xlabel('Samples (n)')
axs[1].set_title('Fs = 200 Hz')

# Plot the FFTs of the original and upsampled signals
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(w*100/(2*np.pi), np.abs(X)/len(X))
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(w_up*200/(2*np.pi), np.abs(X_up)/len(X_up), 'r')
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_title('Fs = 200 Hz')

plt.show()
```

### **The FFT of the obtained signal has replicas from outside the fundamental range, what solution do we have?**

The process of the upsampling by a factor of $I$ and then limited the frequency component in the interval $|\omega_y|=\frac{\pi}{I}$ to avoid the repetitions it is known as **interpolation**.

Interpolation process:

$$v[n]=\begin{cases}x\left[\frac{n}{I}\right],\quad n=rI,\quad r\in \mathbb{Z}\\0,\quad \text{otherwise}\end{cases}$$
$$y[n]=v[n]*h_1[n]\leftrightarrow V(e^{j\omega})H_1(e^{j\omega}),\quad H_1(e^{j\omega})=\begin{cases}I,\quad |\omega|\leq \frac{\pi}{I}\\0,\quad \text{otherwise}\end{cases}$$

![](https://github.com/piero-latorre/DSP/blob/main/VII/image_10.png?raw=true)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define signal
fs = 100
ts = 1/fs
nT = np.arange(0, 1, ts)
nx = np.arange(len(nT))
x = np.sin(2*np.pi*10*nT)

# Interpolate signal
I = 2
x_up = np.interp(np.arange(0, len(x), 1/I), np.arange(0, len(x), 1), x)
nx_up = np.arange(len(x_up))

# Plot original and upsampled signals
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(nx, x, 'b')
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(nx_up, x_up, 'r')
axs[1].set_xlabel('Samples (n)')
axs[1].set_title('Fs = 200 Hz')

# Compute and plot FFTs
X = np.fft.fftshift(np.fft.fft(x))
X_up = np.fft.fftshift(np.fft.fft(x_up))
w = 2*np.pi*nx/len(nx)
w = np.fft.fftshift(w)
w = np.unwrap(w-2*np.pi)
w_up = 2*np.pi*nx_up/len(nx_up)
w_up = np.fft.fftshift(w_up)
w_up = np.unwrap(w_up-2*np.pi)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(w*100/(2*np.pi), np.abs(X)/len(X), 'b')
axs[0].set_title('Fs = 100 Hz')
axs[1].stem(w_up*200/(2*np.pi), np.abs(X_up)/len(X_up), 'r')
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_title('Fs = 200 Hz')

plt.show()
```

###**How can we change the sampling rate by a rational factor?**

It is possible to modify the sampling frequency by a rational factor $F_y=\frac{I}{D} F_x$ following scheme:

* Interpolate $x[n]$ by a factor of $I$

* Decimate the resultant sequence by a factor of $D$

Under this scheme, $H_1(e^{j\omega})$ and $H_D(e^{j\omega})$ are serial systems that affect the input sequence $x[n]$. Hence, it can be replaced by:

$$H(e^{\omega})=\begin{cases}I,\quad|\omega|\leq\min\left\{\frac{\pi}{D};\frac{\pi}{I}\right\} \\ 0,\quad \text{otherwise}\end{cases}$$

![](https://github.com/piero-latorre/DSP/blob/main/VII/image_11.png?raw=true)

![](https://github.com/piero-latorre/DSP/blob/main/VII/image_12.png?raw=true)

Given a rational factor with high values $I$, $D$, it is possible to change the rate sample from a multistage system:

$$T_y=\frac{\prod_{i=0}^N D_i}{\prod_{j=0}^M I_j}T_x$$
**Example**

Considering a $T_y=\frac{63}{130} T_x$, the rate conversion can be found using:

* $D_0=7,D_1=3, D_2=3$

* $I_0 =13,I_1=5, I_2=2 $
