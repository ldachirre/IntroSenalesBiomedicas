# Chapter V: Discrete Fourier Transform

## What is the Discrete-Time Fourier Series?
In discrete-time: a periodic signal $x\left\lbrack n\right\rbrack$ can be related with its Fourier Serie following these equations:
* **Synthesis equation**

$$x[n] = \sum_{k=0}^{N-1} c_k e^{j2\pi kn/N}$$
* **Analysis equation**

$$c_k = \frac{1}{N} \sum_{k = 0}^{N-1} x[n] e^{-j2\pi kn/N}$$
**Example**

Determine the spectum of the periodic signal with $N=4$ :
$$x[n] = \{\underline{1}, 1, 0, 0\}$$


```python
import numpy as np

x = np.array([1, 1, 0, 0]);N = 4
n = np.arange(0, N,dtype="complex64");k = np.arange(0, N,dtype="complex64")
WN = np.exp(-1j * 2 * np.pi / N)
nk = np.outer(n, k)
WNnk = WN ** nk
# print(f"WNnk:{WNnk}")
Xk = x @ WNnk
ck = Xk / N

print(ck)
```

**Example**

Determine the Discrete Fourier Serie of:
$$x[n] = \sin(\omega_o n)$$


```python
import numpy as np

N = 6
n = np.arange(0, N)
x = np.sin(np.pi/3 * n)
k = np.arange(0, N)
WN = np.exp(-1j * 2 * np.pi/N)
nk = np.outer(n, k)
WNnk = WN ** nk
Xk = np.dot(x, WNnk)
ck = Xk/N

print(ck)
```

### What is the Discrete Fourier Transform?
**Is basically taking samples of the Fourier Transform in discrete-time**

Given the causal sequence $x\left\lbrack n\right\rbrack$ with L samples and its spectrum $X\left(e^{j\omega } \right)$, $X_p \left\lbrack k\right\rbrack$ is the infinite duration sequence compound by uniformly spaced $\omega_o =\frac{\;2\pi }{N}$, where $N\in {\mathbb{Z}}^+$.
$$X_p[k]=X(e^{j\omega})|_{\omega=\frac{2\pi k}{N}} = \sum_{n = 0}^{N-1}x[n]e^{-j\frac{2\pi kn}{N}},\quad k =0,\pm 1, \dots$$

Considering that $X\left(e^{j\omega } \right)$ is a $2\pi$ period, then $X_p \left\lbrack k\right\rbrack$ is a sequence with a period equals to N and contains unique values in the interval $k\in \left\lbrace 0,\ldots,N-1\right\rbrace$.

Then, it is possible to obtain an alternative for $X\left\lbrack k\right\rbrack$:
$$X[k] = \begin{cases} X_p[k],\quad k \in \{0, 1, \dots, N-1\} \\ 0,\quad \text{other cases} \end{cases}$$
In other words, $X\left\lbrack k\right\rbrack$ corresponds to a period of $X_p \left\lbrack k\right\rbrack$ in the interval $k\in \left\lbrace 0,\ldots,N-1\right\rbrace$.

From the concept of Discrete-Time Fourier Series, a sequence $x_p \left\lbrack n\right\rbrack$ with period N generates a frequency spectrum of $c_k$ corresponding to an envelope samples of $X\left(e^{j\omega } \right)$.
$$c_k = \mathcal{F} \{x_p[n]\} = \frac{1}{N}X(e^{j\omega})|_{\omega = \frac{2\pi k}{N}}$$
Then, $X_p \left\lbrack k\right\rbrack$ corresponds to the frequency spectrum of $x_p \left\lbrack n\right\rbrack$ scaled by a factor of N:
$$X_p[k] = X(e^{j\omega})|_{\omega = \frac{2\pi k}{N}} = Nc_k$$

### How can we obtain the Inverse Discrete Fourier Transform?
Considering the geometric serie:
$$p[n] = \sum_{r=-\infty}^{+\infty} \delta[n-rN],\quad r\in \mathbb{Z}$$
This imply that $x_p \left\lbrack n\right\rbrack$ corresponds to the version of $x\left\lbrack n\right\rbrack$ with period N.
$$x_p[n] = x[n] * \sum_{r=-\infty}^{+\infty} \delta [n-rN] = \sum_{r=-\infty}^{+\infty}x[n-rN]$$
Then, it is possible to obtain a finite sequence $x\left\lbrack n\right\rbrack$ considering N samples of the spectrum $X\left\lbrack k\right\rbrack$ by using the inverse Discrete Fourier Transform:
$$x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X_p[k]e^{\frac{j2\pi kn}{N}},\quad n\in \{0, 1, \dots, N-1 \}$$
If $N\ge L$, there is not superposition in the componentes of $x_p \left\lbrack n\right\rbrack$ considering the relation between $x\left\lbrack n\right\rbrack$ and $x_p \left\lbrack n\right\rbrack$ is:
$$x_p[n] = x[((n))_N]$$
Where, ${\left(\left(n\right)\right)}_N$ corresponds to a module operation: n mod N. In the same way:
$$x[n] = \begin{cases} x_p[n],\quad n\in\{0, 1, \dots, N-1\} \\ 0,\quad \text{otherwise} \end{cases}$$
If $N<L$, there is superposition in the components, then the original sequence is not preserved.

**Example**
Find  $X\left\lbrack k\right\rbrack$ of $x\left\lbrack n\right\rbrack =\left\lbrace 1,2,3,4\right\rbrace$ and then calculate the inverse.
* $N=5$



```python
import numpy as np
x = [1, 2, 3, 4, 0]
N = 5
n = np.arange(0, N)
k = np.arange(0, N)

WN = np.exp(-1j*2*np.pi/N)
nk = np.outer(n, k)
WNnk = WN ** nk
Xk = np.dot(x, WNnk)
print("Xk:", Xk)

WN = np.exp(-1j*2*np.pi/N)
nk = np.outer(n, k)
WNnk = WN ** (-nk)
xn = np.dot(Xk, WNnk)/N
print("xn:", xn)
```

* $N=3$



```python


x = np.array([1, 2, 3])
N = 3
n = np.arange(N)
k = np.arange(N)
WN = np.exp(-1j * 2 * np.pi / N)
nk = np.outer(n, k)
WNnk = np.power(WN, nk)
Xk = np.dot(x, WNnk)

WN = np.exp(-1j * 2 * np.pi / N)
nk = np.outer(n, k)
WNnk = np.power(WN, -nk)
xn = np.dot(Xk, WNnk) / N

print("Xk =", Xk)
print("xn =", xn)

```

### How can we calculate directly the DFT?
In the previous example, it can be appreciated the condition for an arbitrary sequence. The discrete spectrum $X\left\lbrack k\right\rbrack$ allows the recovering of $x\left\lbrack n\right\rbrack$ only if $N\ge L$. When this condition is not met, the repetition of $x_p \left\lbrack n\right\rbrack$ are superposed generating **time aliasing**.

The relation between $X\left(k\right)$ and $X\left(e^{j\omega } \right)$ using  $L=5$ and $N=10$. The discrete Fourier corresponds to samples of the DTFT spaced by $\Delta \omega =\frac{2\pi }{N}$.


```python

import matplotlib.pyplot as plt

L = 5
n = np.arange(-0.5*(L-1), 0.5*(L-1) + 1)
x = np.cos(0*n)
w = np.linspace(-5*np.pi, 5*np.pi, 501)
X = np.zeros(w.shape, dtype=np.complex128)
for i in range(len(w)):
    for j in range(len(n)):
        X[i] += x[j]*np.exp(-1j*w[i]*n[j])

N = 10
delta_w = 2*np.pi/N
w_DFT = np.arange(-5*np.pi, 5*np.pi + delta_w, delta_w)
DFT = np.zeros(w_DFT.shape, dtype=np.complex128)
for i in range(len(w_DFT)):
    for j in range(len(n)):
        DFT[i] += x[j]*np.exp(-1j*w_DFT[i]*n[j])

plt.figure()
plt.subplot(3, 1, 1)
plt.plot(w, abs(X), "r")
plt.grid(True)
plt.xlabel(r"$\omega$")
plt.ylabel(r"$|X(e^{j\omega})|$")

plt.subplot(3, 1, 2)
plt.stem(w_DFT, abs(DFT))
plt.grid(True)
plt.xlabel("k")
plt.ylabel(r"$|X(k)|$")

plt.subplot(3, 1, 3)
plt.stem(w_DFT, abs(DFT))
plt.grid(True)
plt.plot(w, abs(X))
plt.legend([r"$|X(k)|$", r"$|X(e^{j\omega})|$"])
```

The discrete Fourier corresponds to samples of the DTFT spaced $\Delta \omega =\frac{2\pi }{N}$.


```python
import numpy as np
import matplotlib.pyplot as plt

# taking into consideration that L = 5 and N = 5
L = 5
n = np.arange(-0.5*(L-1), 0.5*(L-1)+1)
x = np.cos(0*n)
w = np.linspace(-10*np.pi, 10*np.pi, 1001)
X = np.zeros_like(w, dtype=np.complex64)
for i in range(len(w)):
    for j in range(len(n)):
        X[i] += x[j]*np.exp(-1j*w[i]*n[j])
N = 5
delta_w = 2*np.pi/N
w_DFT = np.arange(-10*np.pi, 10*np.pi+delta_w, delta_w)
DFT = np.zeros_like(w_DFT, dtype=np.complex64)
for i in range(len(w_DFT)):
    for j in range(len(n)):
        DFT[i] += x[j]*np.exp(-1j*w_DFT[i]*n[j])
plt.figure(figsize=(9,9))
plt.subplot(3, 1, 1)
plt.plot(w, np.abs(X), 'r')
plt.grid(True)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|X(e^{j\omega})|$')
plt.subplot(3, 1, 2)
plt.stem(w_DFT, np.abs(DFT))
plt.grid(True)
plt.xlabel('k')
plt.ylabel(r'$|X(k)|$')
plt.subplot(3, 1, 3)
plt.stem(w_DFT, np.abs(DFT))
plt.grid(True)
plt.plot(w, np.abs(X))
plt.legend([r'$|X(k)|$', r'$|X(e^{j\omega})|$'])
plt.show()

# taking into consideration that L = 10 and N = 5
L = 10
n = np.arange(-0.5*(L-1), 0.5*(L-1)+1)
x = np.cos(0*n)
w = np.linspace(-10*np.pi, 10*np.pi, 1001)
X = np.zeros_like(w, dtype=np.complex64)
for i in range(len(w)):
    for j in range(len(n)):
        X[i] += x[j]*np.exp(-1j*w[i]*n[j])
N = 5
delta_w = 2*np.pi/N
w_DFT = np.arange(-10*np.pi, 10*np.pi+delta_w, delta_w)
DFT = np.zeros_like(w_DFT, dtype=np.complex64)
for i in range(len(w_DFT)):
    for j in range(len(n)):
        DFT[i] += x[j]*np.exp(-1j*w_DFT[i]*n[j])
plt.figure(figsize=(9,9))
plt.subplot(3, 1, 1)
plt.plot(w, np.abs(X), 'r')
plt.grid(True)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$|X(e^{j\omega})|$')
plt.subplot(3, 1, 2)
plt.stem(w_DFT, np.abs(DFT))
plt.grid(True)
plt.xlabel('k')
plt.ylabel(r'$|X(k)|$')
plt.subplot(3, 1, 3)
plt.stem(w_DFT, np.abs(DFT))
plt.grid(True)
plt.plot(w, np.abs(X))
plt.legend([r'$|X(k)|$', r'$|X(e^{j\omega})|$'])
plt.show()

```

It is possible to express the DFT by using matrix-vector operation:
$$W_N = e^{-j2\pi \frac{1}{N}}$$
$$\therefore X[k] = \sum_{n=0}^{N-1} x[n] W_N^{kn},\quad k \in \{0, 1, \dots, N-1\}$$
$$x[n] = \frac{1}{N}\sum_{k=0}^{N-1} X[k]W_N^{-kn},\quad n \in \{0, 1, \dots, N-1\}$$


**Example**
Calculate the 4-point DFT of $x\left\lbrack n\right\rbrack =\left\lbrace 0,1,2,3\right\rbrace$.


```python


x = np.array([0, 1, 2, 3])
N = 4
n = np.arange(N)
k = np.arange(N)
WN = np.exp(-1j*2*np.pi/N)
nk = np.outer(n, k)
WNnk = WN ** nk
Xk = np.dot(x, WNnk)
print(Xk)

```

### What are the Properties of the DFT?
**Circular shift**

The DFT with N points corresponds to a sequence $x\left\lbrack n\right\rbrack$ with length of $L\le N$ equivalent to the Fourier serie (by a N factor) of the periodic sequence $x_p \left\lbrack n\right\rbrack$ with N of the periodic:
$$x_p[n] = \sum_{l=-\infty}^\infty x[n-lN]$$
In the same way, the DFT with N points of $\hat{x} \left\lbrack n\right\rbrack =x\left\lbrack {\left(\left(n-k\right)\right)}_N \right\rbrack$ (shifted displacement with k element in a circular shape) corresponding to a Fourier serie (by a N factor) of the periodic sequence:
$$\hat{x}[n] = x_p[n-k] = \sum_{l=-\infty}^\infty x[n-k-lN]$$
$$\hat{x}[n] = \begin{cases} \hat{x}_p [n],\quad n \in \{0, \dots, N-1\} \\ 0, \quad \text{otherwise} \end{cases} = x[((n-k))_N]_m, \quad n\in \{0,\dots,N-1\}$$
Finally, $x\left\lbrack n\right\rbrack$ and $\hat{x} \left\lbrack n\right\rbrack$ are composed by the same elements but circular shifted.


```python

# Perform circular shifting for x = [1, 2, 3, 4] with 2 samples
x = np.array([1, 2, 3, 4])
n = np.arange(4)
y = np.roll(x, 2)

plt.figure()
plt.subplot(2, 1, 1)
plt.stem(n, x, 'b')
plt.title('$x_p[n]$')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(n, y, 'r')
plt.xlabel('Samples (n)')
plt.title('$x_p[n-2]$')
plt.grid(True)

plt.show()
```

### Frequency product of the DFT and circular shift

Considering the finite duration sequence lower than $N\left\lbrace x\left\lbrack n\right\rbrack ,h\left\lbrack n\right\rbrack \right\rbrace$ with DTFT $\left\lbrace X\left(e^{j\omega } \right),H\left(e^{j\omega } \right)\right\rbrace$ and DFT $\left\lbrace X\left\lbrack k\right\rbrack ,H\left\lbrack k\right\rbrack \right\rbrace$, respectively. Additionally,$\left\lbrace X_p \left\lbrack k\right\rbrack ,H_p \left\lbrack k\right\rbrack \right\rbrace$  corresponds to the samples of $\left\lbrace X\left(e^{j\omega } \right),H\left(e^{j\omega } \right)\right\rbrace$ and the sequence $\left\lbrace x_p \left\lbrack n\right\rbrack ,h_p \left\lbrack n\right\rbrack \right\rbrace$ has period N, there is a difference by using the multiplication in the continuous frequency domain $\left\lbrace X\left(e^{j\omega } \right),H\left(e^{j\omega } \right)\right\rbrace$ and their discrete version $\left\lbrace X_p \left\lbrack k\right\rbrack ,H_p \left\lbrack k\right\rbrack \right\rbrace$. From the convolution property, the new relation corresponds to:

![image.png](https://github.com/piero-latorre/DSP/blob/main/V/image_2.png?raw=true)

In general, for a LTI system with an impulse response $h\left\lbrack n\right\rbrack$ and a input sequence $x\left\lbrack n\right\rbrack$, the output could be calculated by the product of the discrete Fourier transform with number of samples $N\ge P+Q-1$ , where P and Q corresponds to the size of the impulse response and the input, respectively:

![image.png](https://github.com/piero-latorre/DSP/blob/main/V/image_3.png?raw=true)

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(100)
x = np.sin(np.pi/3 * n)
h = np.power(0.5, n)
y = np.convolve(x, h, mode='full')

N = len(y)
n = np.arange(N)
k = np.arange(N)
WN = np.exp(-1j*2*np.pi/N)
nk = np.outer(n, k)
WNnk = np.power(WN, nk)
Yk = np.dot(y, WNnk)

N = 2*len(x) - 1
n = np.arange(N)
k = np.arange(N)
WN = np.exp(-1j*2*np.pi/N)
nk = np.outer(n, k)
WNnk = np.power(WN, nk)
Xk = np.hstack((x, np.zeros(N-len(x)))) @ WNnk
Hk = np.hstack((h, np.zeros(N-len(h)))) @ WNnk
Y = Xk * Hk

plt.figure()
plt.subplot(2, 1, 1)
plt.stem(np.abs(Yk), use_line_collection=True)
plt.title('DFT of y[n] with circular convolution')
plt.subplot(2, 1, 2)
plt.stem(np.abs(Y), use_line_collection=True)
plt.title('DFT of y[n] from DFT product')
plt.xlabel('Samples (k)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(100)
x = np.sin(np.pi/3 * n)
h = np.power(0.5, n)
y1 = np.convolve(x, h, mode='full')
y2 = np.convolve(h, x, mode='full')

plt.figure()
plt.subplot(2, 1, 1)
plt.stem(np.arange(len(y1)), y1, 'b')
plt.title('x[n] * h[n]')

plt.subplot(2, 1, 2)
plt.stem(np.arange(len(y2)), y2, 'r')
plt.title('h[n] * x[n]')

plt.show()
```
